```python
# Bert Layers
text_input = tf.keras.Input(shape=(), dtype=tf.string, name='text')
bert_preprocess = hub.KerasLayer(preprocess_url, name='preprocessing')
preprocessed_text = bert_preprocess(text_input)
bert_encoder = hub.KerasLayer(encoder_url, 
                              trainable=True, 
                              name='BERT_encoder')
outputs = bert_encoder(preprocessed_text)

# Neural network layers
l = tf.keras.layers.Dropout(0.1)(outputs['pooled_output'])
l = tf.keras.layers.Dense(num_classes, activation='softmax', name='output')(l)

# Construct final model
classifier_model = tf.keras.Model(inputs=[text_input], outputs=[l])
```

```python
bert_raw_result = classifier_model(tf.constant(test_text))
```

```python
bert_raw_result
```

```python
tf.keras.utils.plot_model(classifier_model)
```

```python
classifier_model.compile(optimizer=tf.keras.optimizers.AdamW(), metrics=['accuracy'], loss=tf.keras.losses.CategoricalCrossentropy())
```
